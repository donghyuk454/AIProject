## Exercise 1

Create a logger that would print the following information in the followed format.  
Your logger name is your last name.

```
{month}-{day} {hour}:{minute}:{second} - {logging level} - {logger name} - {message}
```

````python
import logging
````

````python
logger = logging.getLogger('Dong-hyuk')

logger.setLevel(logging.INFO)

formatter = logging.Formatter(
    fmt='%(asctime)s - %(levelname)s - %(name)s - %(message)s',
    datefmt='%m-%d %H:%M:%S'
)
streamhandler = logging.StreamHandler()
from datetime import datetime

streamhandler.setFormatter(formatter)

logger.addHandler(streamhandler)
````

````python
logger.info('Nicely done.')
````

## Exercise 2

Draw two different plots with Iris dataset:
1. Violin plot between target and sepal width
2. Scatter plot between petal width and petal length with targets colored

````python
%matplotlib inline

import seaborn as sns

iris = sns.load_dataset('iris')
````

````python
# Write your code
sns.violinplot(x='sepal_width', y='species', data=iris)
````

````python
# Write your code
sns.scatterplot(x="petal_width", y="petal_length", hue="species", data=iris)
````
